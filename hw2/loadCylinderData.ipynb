```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import svd
import scipy.io # for importing file
```

```python
# Load the data from the .mat file
# This function will load the data from the .mat file and create variables in the global namespace
# with the same name as the keys in the .mat file.
def load_mat_as_variables(file_path):
    mat_data = scipy.io.loadmat(file_path)
    for key, value in mat_data.items():
        if not key.startswith("__"): # Avoid loading metadata keys
            globals()[key] = value

file_path = 'CYLINDER_ALL.mat'
load_mat_as_variables(file_path)
```
